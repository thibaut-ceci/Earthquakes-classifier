# Explore the database

Here, we want to observe the distribution of data in the database.

```python
import catalog

catalog.display_parameters()
```

## Load the database

```python
database = catalog.load("data/new_database.pickle")

catalog.open_plot(database, pos_number=0.035, xlim=26000)
```

## Distribution of "type" column

```python
catalog.see_number_distribution_number(database["type"], "Type", 15000)
```

## Distribution of "mag" (magnitude) column

```python
catalog.see_number_distribution(database["mag"], "Magnitude")
```
