# Explore the database

Here, we want to observe the distribution of data in the database.

```python
import catalog

catalog.display_parameters()
```

## Load the database

```python
database = catalog.load("data/new_database.pickle")

catalog.open_plot(database, pos_number=0.040, xlim=26000)
```

## Distribution of "type" column

```python
import matplotlib.pyplot as plt

result = plt.hist(database["type"], orientation="horizontal", bins=10)
plt.xlabel("Number of events")
plt.ylabel("Type")

for index, value in enumerate(result[0]):
    plt.text(value, index, str(int(value)))

plt.tight_layout()
plt.show()
```

```python
catalog.see_number_distribution_number(database["type"], "Type", 15000)
```
