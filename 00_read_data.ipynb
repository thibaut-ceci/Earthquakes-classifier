```python
import pandas as pd

import catalog

catalog.display_parameters()
```

```python
EQ = pd.read_csv("data/query.csv")
NONEQ = pd.read_csv("data/NONEQ.csv")
```

```python
EQ = EQ[EQ["mag"] > 5.3]
print("Séisme ", len(EQ))
```

```python
NONEQ = NONEQ[NONEQ["mag"] > 2.7]
print("Pas séisme ", len(NONEQ))
dataframe = pd.concat([EQ, NONEQ], ignore_index=True)
```

```python
print("Taille catalogue ", len(dataframe))
```

```python
colonne = ['depth', 'nst', 'gap', 'dmin', 'rms', 'net', 'magType', 'id', 'horizontalError', 'depthError', 'magError', 'magNst', 'locationSource', 'magSource', 'updated', 'status']
dataframe.drop(colonne, axis=1, inplace=True)
```
