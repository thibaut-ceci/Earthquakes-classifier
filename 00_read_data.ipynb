# What is "Earthquakes classifier"?

Earthquakes classifier is a project which aims to create a machine learning algorithm to classify seismic signals from earthquakes or others.
A database based on 1000 seismic signals from earthquakes and 1000 from other sources will be used.

The data comes from USGS : https://earthquake.usgs.gov/earthquakes/search/

---

# Read data

## First, let's load the librairies and the two databases

```python
## Librairies
import pandas as pd

## Handmade librairies
import catalog

## Change catalog parameters
catalog.display_parameters() ## To see the entire catalog when it is called
```

```python
## Load the database with the seismic signal from earthquakes
Database_EQ = pd.read_csv("data/database_eq.csv")

## Load the database with the seismic signal from other sources 
Database_not_EQ = pd.read_csv("data/database_not_eq.csv")
```

## Now, let's open the two databases

```python
catalog.open_plot(Database_EQ, pos_number=0.15, xlim=15000) 
```

```python
catalog.open_plot(Database_not_EQ, pos_number=0.15, xlim=21000)
```

## Reduce the databases

We can see that we have 12906 seismic signals from earthquake and 18221 seismic signals from others sources. In this study, we will work with only 1000 seismic signals from earthquakes and 1000 seismic signals from other sources. Let's reduce the databases :

```python
## Keep the events from earthquakes with a magnitude of 5.3 or greater
Database_EQ = Database_EQ[Database_EQ["mag"] > 4]
print("There are", len(Database_EQ), "seismic signals from earthquakes")

## Keep the events from other sources with a magnitude of 2.7 or greater
Database_not_EQ = Database_not_EQ[Database_not_EQ["mag"] > 1.3]
print("There are", len(Database_not_EQ), "seismic signals from other sources")

## Merge the databases
dataframe = pd.concat([Database_EQ, Database_not_EQ], ignore_index=True)

## See the size of the new database
print("The size of the new database is", len(dataframe))

## Save the new database
dataframe.to_pickle("data/new_database.pickle")
```
